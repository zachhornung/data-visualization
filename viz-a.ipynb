# Did Precipitation levels affect bombing operations in world war 2?
* This notebook calculates the average precipitation for all weather stations for all months during the years 1939-1945.
* Notebook viz-b calculates the average bombing missions for all months during the same time period.

```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib as plt
%matplotlib inline
```

```python
weather = pd.read_csv('weather.csv')
```

```python
weather['month'] = pd.DatetimeIndex(weather['Date']).month
weather['year'] = pd.DatetimeIndex(weather['Date']).year
```

```python
weather['Precip'] = pd.to_numeric(weather['Precip'],errors='coerce')
x = weather.groupby(["year", "month"], as_index=False)
monthly_averages = x.aggregate({"Precip":np.mean})

```

```python

sns.relplot(x="year",y="Precip", hue="month", data=monthly_averages, kind='line')
```
